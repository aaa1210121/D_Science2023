### **Traversal with **'while'** Loop**

```python
""" while loop """
index = 0
fruit = "apple and banana"
while index < len(fruit):
    letter = fruit[index]
    print(letter)
    index = index + 1
```

### **Traversal with **'for'** Loop**

```python
""" for loop """
fruit = "apple and banana"
for letter in fruit:
    print(letter)
```

### **Reading a txt**

```python
# Get your working directory
path = os.getcwd()
# Join dir path with file path
file_path = os.path.join(path, 'example.txt')
# Open and read file

''' PLEASE note that Python_Basics.ipynb and example.txt in the same directory'''
with open(file_path, 'r', encoding = 'UTF-8') as f:
    for line in f.readlines():
        print(line)
```

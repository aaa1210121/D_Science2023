## **<span style='color:#EA7500	'>Wirting a CSV</span>**

```python
with open('employee_file.csv', mode='w') as employee_file:
    employee_writer = csv.writer(employee_file, delimiter=',')
    employee_writer.writerow(['John Smith', 'Accounting', 'November'])
    employee_writer.writerow(['Erica Meyers', 'IT', 'March'])

with open('employee_file.csv', 'r') as csv_file:
    csv_file = csv.reader(csv_file, delimiter=',')

    for inf in csv_file:
        print(inf[0], inf[1], inf[2])
```

## **<span style='color:#EA7500	'>Writing a CSV with Header</span>**


```python
with open('employee_file.csv', mode='w') as employee_file:
    fieldnames = ['Name', 'Occupation', 'Birthday']
    employee_writer = csv.DictWriter(employee_file, delimiter=',', fieldnames=fieldnames)

    employee_writer.writeheader()
    employee_writer.writerow({'Name':'John Smith', 'Occupation':'Accounting', 'Birthday':'November'})
    employee_writer.writerow({'Name':'Erica Meyers', 'Occupation':'IT', 'Birthday':'March'})

with open('employee_file.csv', 'r') as csv_file:
    csv_file = csv.reader(csv_file, delimiter=',')

    for inf in csv_file:
        print(inf[0], inf[1], inf[2])
```

### **<span style='color:#EA7500	'>Array Slicing: 1D</span>**
##### **array[start:stop:step] default values: start=0, stop=size of dimension, step=1**

```python
x = np.arange(10)
x[:5] # first five elements
x[5:] # elements after index 5
x[4:7] # middle sub-array
x[::2] # every other element
x[::-1] # all elements, reversed
```

### **<span style='color:#EA7500	'>Array Slicing: 2D</span>**


```python
x = np.arange(12).reshape(3,4)
print(f'x\n{x}\n\nx[:2][:3]\n{x[:2, :3]}\n\nx[:3][::2]\n{x[:3, ::2]}')
print(f'\nx[:-1][:-1]\n{x[::-1, ::-1]}')
```

### **<span style='color:#EA7500	'>Array aliasing</span>**


```python
x1 = np.random.randint(10, size=(3, 4))
x2 = x1[:2, :3]
print(f'{x1}\n-----------\n{x2}\n')
x2[0, 0] = 99
print(f'{x1}\n-----------\n{x2}')
```

### **<span style='color:#EA7500	'>Creating Copies of Arrays</span>**


```python
x1 = np.random.randint(10, size=(3, 4))
x2 = x1[:2, :2].copy()
print(f'{x1}\n-----------\n{x2}\n')
x2[0, 0] = 99
print(f'{x1}\n-----------\n{x2}')
```

### **<span style='color:#EA7500	'>Array Arithmatic</span>**


```python
x = np.arange(4)
print("x =", x)
print("x + 5 =", x + 5)
print("x - 5 =", x - 5)
print("x * 2 =", x * 2)
print("x / 2 =", x / 2)
print("x // 2 =", x // 2)
```

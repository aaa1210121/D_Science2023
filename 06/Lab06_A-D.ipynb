```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("sales_data.csv")


with plt.xkcd():
    plt.figure(figsize= (9,6))
    fig, ax = plt.subplots(2, 1, constrained_layout=True)
    # fig.text(0.01, 0.5, 'Sales units  in number', va='center', rotation='vertical')
    ax[0].set_title("Sales data of a Bathingsoap" , fontsize = 24)
    ax[0].tick_params(axis="x", labelsize=18)
    ax[0].set_xticks(df["month_number"])
    ax[0].set_xticklabels([])
    ax[0].plot(df["month_number"], df["bathingsoap"], "-ok" , markersize = 10, linewidth = 4)

    ax[1].set_title("Sales data of a facewash", fontsize = 24)
    ax[1].set_xticks(df["month_number"])
    ax[1].tick_params(axis="x", labelsize=18)
    ax[1].set_xlim(0, 13)
    ax[1].set_xlabel("Month Number" , fontsize =18)
    ax[1].plot(df["month_number"], df["facewash"], "-ok", markersize = 10 , linewidth = 4)
    # fig.tight_layout()
    fig.supylabel("Sales units in number", fontsize = 18)
    plt.show()
```

# Lab06_B

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_excel("Scatter.xlsx")
fig = plt.figure(figsize=(6, 4))
colors = {0: "#FF0033", 1: "#FFFF66", 2: "#99CCFF"}
m = ["^", "o", "s"]
label = ["c/c", "C/c", "C/C"]

ax3 = plt.subplot2grid(shape=(3, 3), loc=(0, 1), rowspan=2, colspan=2)
ax3.set_title("Scatter Plot", fontsize=12)
ax3.set_ylabel("PC2", fontsize=10)
ax3.set_xlabel("PC1", fontsize=10)
ax3.set_xticks(np.arange(-400, 600, 200))
ax3.set_yticks(np.arange(-400, 600, 200))
ax3.set_ylim(-400, 400)
ax3.set_xlim(-400, 400)
ax3.xaxis.set_ticks_position("both")
ax3.yaxis.set_ticks_position("both")
ax3.tick_params(direction="in")

for i, color in colors.items():
    x = df.loc[df["Genotype"] == i, "PC1"]
    y = df.loc[df["Genotype"] == i, "PC2"]
    ax3.scatter(
        x, y, c=color, marker=m[i], clip_on=False, label=label[i], edgecolor="black"
    )

ax2 = plt.subplot2grid(shape=(3, 3), loc=(2, 0))
ax2.set_xticks([])
ax2.set_yticks([])
label = ["c/c", "C/c", "c/C"]
handles, labels = ax3.get_legend_handles_labels()
ax2.legend(handles, labels, loc="center", edgecolor="None" ,frameon = False)

colors = ["#FF0033", "#FFFF66", "#99CCFF"]
x1 = df.loc[df["Genotype"] == 0, "PC1"]
x2 = df.loc[df["Genotype"] == 1, "PC1"]
x3 = df.loc[df["Genotype"] == 2, "PC1"]
ax4 = plt.subplot2grid(shape=(3, 3), loc=(2, 1), colspan=2)
ax4.set_ylabel("Frequency", fontsize=10)
ax4.set_xlabel("PC1", fontsize=10)
ax4.set_xticks(np.arange(-400, 600, 200))
ax4.set_yticks(np.arange(0, 15, 5))
ax4.set_ylim(0, 10)
ax4.set_xlim(-400, 400)
ax4.hist(
    [x1, x2, x3], bins=10, stacked=True, rwidth=0.8, color=colors, edgecolor="black"
)
ax4.xaxis.set_ticks_position("both")
ax4.yaxis.set_ticks_position("both")
ax4.tick_params(direction="in")
ax4.tick_params(axis="x", which="both", bottom=False)

y1 = df.loc[df["Genotype"] == 0, "PC2"]
y2 = df.loc[df["Genotype"] == 1, "PC2"]
y3 = df.loc[df["Genotype"] == 2, "PC2"]
ax1 = plt.subplot2grid(shape=(3, 3), loc=(0, 0), rowspan=2)
ax1.set_ylabel("PC2", fontsize=10)
ax1.set_xlabel("Frequency", fontsize=10)
ax1.hist(
    [y1, y2, y3],
    bins=10,
    stacked=True,
    rwidth=0.8,
    color=colors,
    orientation="horizontal",
    edgecolor="black",
)
ax1.set_ylim(-400, 400)
ax1.set_yticks(np.arange(-400, 600, 200))
ax1.set_xlim(0, 10)
ax1.xaxis.set_ticks_position("both")
ax1.yaxis.set_ticks_position("both")
ax1.tick_params(direction="in")
ax1.tick_params(axis="y", which="both", left=False)

fig.tight_layout()
plt.show()

```

# Lab06_C

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv("Score.csv")
x = df["Round"].unique()

fig, ax = plt.subplots(figsize=(11, 6))
ax.set_title("Points", fontsize=16)
ax.spines["bottom"].set_visible(False)
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.grid(axis="x", color="#999999")
ax.xaxis.tick_top()
ax.xaxis.label.set_color("#999999")
ax.tick_params(axis="x", colors="#CCCCCC")
ax.tick_params(axis="y", direction="in", colors="k")
ax.set_xlim(0, 170)

d = np.array(df.loc[df["Team"] == "D", "Score"])
k = np.array(df.loc[df["Team"] == "K", "Score"])
p = np.array(df.loc[df["Team"] == "P", "Score"])
p = np.append(p, [0, 0])

B_d = ax.barh(x, d, color="#CCFF99", left=0.2, label="Team D", zorder=100)
B_k = ax.barh(x, k, color="#66CCFF", left=d + 0.5, label="Team K", zorder=100)
B_P = ax.barh(
    x,
    p,
    color="#FFCC00",
    left=k + d + 1,
    label="Team P",
    zorder=100,
)  # zorder=100
labels = ["Team k", "Team D", "Team p"]
colors = ["Blue", "Green", "Yellow"]


for (
    i,
    j,
    name_d,
    name_k,
) in zip(
    B_d.patches,
    B_k.patches,
    df.loc[df["Team"] == "D", "Leader"],
    df.loc[df["Team"] == "K", "Leader"],
):
    wd_d = i.get_width() / 2
    wd_k = j.get_width() / 2 + i.get_width()
    ax.text(wd_d, i.get_y() + 0.35, name_d, zorder=1000, fontsize=14)
    ax.text(wd_k, i.get_y() + 0.35, name_k, zorder=1000, fontsize=14)
name_p = df.loc[df["Team"] == "P", "Leader"].values[0]
ax.text(134.2, 14.95, name_p, zorder=1000, fontsize=14)
ax.text(110.2, 13.95, name_p, zorder=1000, fontsize=14)
ax.text(127.2, 12.95, name_p, zorder=1000, fontsize=14)
ax.tick_params(axis="y", which="both", left=False)

lab = ["Team K", "Team D", "Team P"]
ax.legend(
    handles=[B_k, B_d, B_P],
    labels=lab,
    loc="lower right",
    bbox_to_anchor=(0.97, 0.03),
    handlelength=1,
    handleheight=1.2,
    fontsize=14,
)

plt.show()
```

# Lab06_D

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 5, 0.01)
y = np.arange(0, 5, 0.01)
x, y = np.meshgrid(x, y)
p = 0
f_x_y = (1 / (2 * np.pi * (np.sqrt(1 - p**2)))) * np.exp(
    -1
    / 2
    * (1 - p**2)
    * ((x - 2.5) ** 2 + (y - 2.5) ** 2 - 2 * p * (x - 2.5) * (y - 2.5))
)

fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")


ax.plot_surface(x, y, f_x_y, cmap="viridis")
ax.set_xlabel("X axis")
ax.set_ylabel("Y axis")
ax.set_zlabel("Z axis")

fig.tight_layout()
plt.show()

```

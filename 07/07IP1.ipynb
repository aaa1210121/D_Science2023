```python
import skimage
from skimage import io, img_as_float, color
import matplotlib.pyplot as plt

balloon = io.imread("monkey_islan.jpg")
balloon = img_as_float(balloon)
plt.figure(figsize=(10, 8))

balloon_gray1 = color.rgb2gray(balloon)
balloon_gray2 = balloon.mean(axis=2)
balloon_gray3 = 0.2989 * balloon[:, :, 0] + 0.5870 * balloon[:, :, 1] + 0.1140 * balloon[:, :, 2]

plt.subplot(1, 4, 1)
plt.imshow(balloon)
plt.title("Color Image")

plt.subplot(1, 4, 2)
plt.imshow(balloon_gray1, cmap="gray")
plt.title("Grayscale (rgb2gray)")

plt.subplot(1, 4, 3)
plt.imshow(balloon_gray2, cmap="gray")
plt.title("Grayscale (mean)")

plt.subplot(1, 4, 4)
plt.imshow(balloon_gray3, cmap="gray")
plt.title("Grayscale (custom)")

plt.show()
```

## **<span style='color:#EA7500	'>Integers can overflow</span>**

```python
from skimage import img_as_float

camera_float = img_as_float(camera)
print(camera.max(), camera_float.max())
camera_float_15 = camera_float*1.5
print(camera_float_15.max())

plt.subplot(1, 2, 1)
plt.imshow(camera_float_15, cmap="gray")
plt.subplot(1, 2, 2)
plt.imshow(camera_float_15, vmax=1, cmap="gray")
plt.setp(plt.gcf().get_axes(), xticks=[], yticks=[])
plt.show()
```

## **<span style='color:#EA7500	'>Image Multiplication</span>**

## **<span style='color:#EA7500	'>Image Addition</span>**

```python
from skimage import transform

rice = io.imread("rice.jpg")
rice = transform.resize(rice, (512,512))
addition = rice + camera_float

plt.figure(figsize=(20,20))
plt.subplot(1, 3, 1)
plt.imshow(rice, cmap="gray")
plt.subplot(1, 3, 2)
plt.imshow(camera_float, cmap = "gray")
plt.subplot(1, 3, 3)
plt.imshow(addition, cmap="gray")
plt.setp(plt.gcf().get_axes(), xticks=[], yticks=[])
plt.show()
```

## **<span style='color:#EA7500	'>Image Subtraction</span>**

```python
road = io.imread("highway1.jpg")
car_on_road = io.imread("highway2.jpg")


road = img_as_float(road)
car_on_road = img_as_float(car_on_road)
car = car_on_road-road


car_mask = car > 0.3

plt.figure(figsize=(10,8))
plt.subplot(2, 2, 1)
plt.imshow(car_on_road, cmap="gray")
plt.subplot(2, 2, 2)
plt.imshow(road, cmap="gray")
plt.subplot(2, 2, 3)
plt.imshow(car, cmap="gray")
plt.subplot(2, 2, 4)
plt.imshow(car_mask, cmap="gray")
plt.show()
```

## **<span style='color:#EA7500	'>Image Rotation</span>**

```python
from skimage import transform, data
camera = data.camera()
rotate_camera = transform.rotate(camera, 30, resize=False, order=1)
rotate_camera_resize = transform.rotate(camera, 30, resize=True, order=1)
# order: https://scikit-image.org/docs/dev/api/skimage.transform.html#skimage.transform.warp

plt.figure(figsize=(10,8))
plt.subplot(1, 3, 1)
plt.imshow(camera, cmap="gray")
plt.subplot(1, 3, 2)
plt.imshow(rotate_camera, cmap="gray")
plt.subplot(1, 3, 3)
plt.imshow(rotate_camera_resize, cmap="gray")
plt.show()
```

## **<span style='color:#EA7500	'>transform.AffineTransform()</span>**

```python
tform = transform.AffineTransform(
            scale=(2, 4), 
            rotation=-np.pi/3, 
            shear=np.pi/6, 
            translation=(-400, 200)
        )
print(tform.params)
tf_camera = transform.warp(camera, tform)

plt.imshow(tf_camera, cmap='gray')
plt.show()
```
